```python
from NuPyCEE import sygma as s
from NuPyCEE import omega as o
import matplotlib.pyplot as plt
```

```python
s1=s.sygma(mgal=1e11,dt=1e7,tend=1e11,imf_type='kroupa',imf_bdys=[1,30],iniZ=0.02,hardsetZ=0.02)
s1.plot_totmasses(source='agb')
s1.plot_totmasses(source='massive')
s1.plot_totmasses(source='all')
```

```python
s2=s.sygma(mgal=1e11,dt=1e7,tend=1.5e10,imf_type='kroupa',imf_bdys=[1,30],iniZ=0.02,hardsetZ=0.02)
s2.plot_totmasses(source='agb')
s2.plot_totmasses(source='massive')
s2.plot_totmasses(source='all')
```

```python
s3=s.sygma(mgal=1e11,dt=1e7,tend=1.5e10,imf_type='galimf',imf_bdys=[1,30],iniZ=0.02,hardsetZ=0.02)
s3.plot_totmasses(source='agb')
s3.plot_totmasses(source='massive')
s3.plot_totmasses(source='all')
```

```python
o1=o.omega(in_out_control=True,cte_sfr=1.0,inflow_rate=1.0,imf_type='kroupa')
o2=o.omega(in_out_control=True,cte_sfr=1.0,inflow_rate=1.0,imf_type='galimf')
o1.plot_spectro(xaxis='age',yaxis='[C/H]',color='b',label='kroupa')
o2.plot_spectro(xaxis='age',yaxis='[C/H]',color='r',label='galimf')
```

```python
o1.plot_spectro(xaxis='age',yaxis='[N/H]',color='b',label='kroupa')
o2.plot_spectro(xaxis='age',yaxis='[N/H]',color='r',label='galimf')
```
