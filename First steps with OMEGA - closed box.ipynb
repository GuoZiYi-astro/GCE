```python
# package of python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from NuPyCEE import omega as o
```

## Initial mass of the gas reservoir

```python
# set constant star formation rate as 1 Msun/yr
o_cons_sfr_10=o.omega(cte_sfr=1.0,mgal=1e10)
o_cons_sfr_11=o.omega(cte_sfr=1.0,mgal=1e11)
o_cons_sfr_12=o.omega(cte_sfr=1.0,mgal=1e12)
```

```python
%matplotlib nbagg
o_cons_sfr_10.plot_totmasses(color='r',label='M=1e10',marker='o')
o_cons_sfr_11.plot_totmasses(color='g',label='M=1e11',marker='^')
o_cons_sfr_12.plot_totmasses(color='b',label='M=1e12',marker='s')
plt.ylim(1e9,2e12)
```

```python
# the evolution of Fe abundance 
%matplotlib nbagg
y='[Fe/H]'
o_cons_sfr_10.plot_spectro(yaxis=y,color='r',label='M=1e10',shape='-')
o_cons_sfr_11.plot_spectro(yaxis=y,color='g',label='M=1e11',shape='--')
o_cons_sfr_12.plot_spectro(yaxis=y,color='b',label='M=1e12',shape='-.')
```

```python
# the evolution of Si to Fe abundance ratio
%matplotlib nbagg
x='age'
y='[Si/Fe]'
o_cons_sfr_10.plot_spectro(xaxis=x,yaxis=y,color='r',label='M=1e10')
o_cons_sfr_11.plot_spectro(xaxis=x,yaxis=y,color='g',label='M=1e11')
o_cons_sfr_12.plot_spectro(xaxis=x,yaxis=y,color='b',label='M=1e12')
```

```python
# the ratio of Si to Fe as a function of the abundance of Fe
%matplotlib nbagg
y='[Si/Fe]'
x='[Fe/H]'
o_cons_sfr_10.plot_spectro(xaxis=x,yaxis=y,color='r',label='M=1e10')
o_cons_sfr_11.plot_spectro(xaxis=x,yaxis=y,color='g',label='M=1e11')
o_cons_sfr_12.plot_spectro(xaxis=x,yaxis=y,color='b',label='M=1e12')
```

```python
# the ratio of Si to Fe as a function of the abundance of Fe
%matplotlib nbagg
y='[Si/Fe]'
x='[Fe/H]'
o_cons_sfr_10.plot_spectro(xaxis=x,yaxis=y,color='r',label='M=1e10')
o_cons_sfr_11.plot_spectro(xaxis=x,yaxis=y,color='g',label='M=1e11')
o_cons_sfr_12.plot_spectro(xaxis=x,yaxis=y,color='b',label='M=1e12')
plt.ylim(-0.2,2.5)
```

## The number of Type Ia supernovae

```python
# set the constant star formation rate and different ratio of Type Ia supernovae
o_cons_low_1a=o.omega(cte_sfr=1.0,mgal=1e10,nb_1a_per_m=1e-4)
o_cons_1a=o.omega(cte_sfr=1.0,mgal=1e10,nb_1a_per_m=1e-3)
o_cons_high_1a=o.omega(cte_sfr=1.0,mgal=1e10,nb_1a_per_m=1e-2)
```

```python
# the ratio of Si to Fe as a function of the abundance of Fe
%matplotlib nbagg
x='[Fe/H]'
y='[Si/Fe]'
o_cons_low_1a.plot_spectro(xaxis=x,yaxis=y,color='r',label='SN Ia ratio=1e-4')
o_cons_1a.plot_spectro(xaxis=x,yaxis=y,color='g',label='SN Ia ratio=1e-3')
o_cons_high_1a.plot_spectro(xaxis=x,yaxis=y,color='b',label='SN Ia ratio=1e-2')
```
