```python
from NuPyCEE import omega as o
from NuPyCEE import sygma as s
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

```python
s1=s.sygma(mgal=1e9)
s2=s.sygma(mgal=1e10)
s3=s.sygma(mgal=1e11)
```

```python
y='[Fe/H]'
s1.plot_spectro(yaxis=y,color='k',label='m=1e9')
s2.plot_spectro(yaxis=y,color='m',label='m=1e10')
s3.plot_spectro(yaxis=y,color='r',label='m=1e11')
```

I try to set different mgal in SYGMA to see if the mgal will change the abundance of Fe, but seems that the mgal do not effect the abundance of Fe. What change in OMEGA still not very clear.

```python
o1=o.omega(mgal=1e10)
o2=o.omega(mgal=1e11)
o3=o.omega(mgal=1e12)
```

```python
y='Fe'
o1.plot_massfrac(yaxis=y,color='k',label='M=1e10')
o2.plot_massfrac(yaxis=y,color='r',label='M=1e11')
o3.plot_massfrac(yaxis=y,color='m',label='M=1e12')
```

```python
y='Fe'
o1.plot_mass(specie=y,color='k',label='M=1e10')
o2.plot_mass(specie=y,color='r',label='M=1e11')
o3.plot_mass(specie=y,color='m',label='M=1e12')
```

```python
y='Fe'
xy_1=o1.plot_mass(specie=y,color='k',return_x_y=True)
xy_2=o2.plot_mass(specie=y,color='r',return_x_y=True)
xy_3=o3.plot_mass(specie=y,color='m',return_x_y=True)

plt.plot(xy_1[0],xy_1[1],color='k',label='M=1e10',linewidth=1.5)
plt.plot(xy_2[0],xy_2[1],color='r',label='M=1e11',linewidth=1.5)
plt.plot(xy_3[0],xy_3[1],color='m',label='M=1e12',linewidth=1.5)
plt.legend(loc='center left', bbox_to_anchor=(1.01, 0.5), markerscale=0.8, fontsize=14)
# plt.xscale=('log')
```

After checking the default sfr, we believe the difference is because of the cte_sfr. So if we change the cte_sfr as well, the [Fe/H] should be the same.

```python
o4=o.omega(mgal=1e10,cte_sfr=1)
o5=o.omega(mgal=1e11,cte_sfr=10)
o6=o.omega(mgal=1e12,cte_sfr=100)
```

```python
y='[Fe/H]'
o4.plot_spectro(yaxis=y,color='k',label='M=1e10,cte_sfr=1')
o5.plot_spectro(yaxis=y,color='r',label='M=1e11,cte_sfr=10')
o6.plot_spectro(yaxis=y,color='m',label='M=1e12,cte_sfr=100')
```
