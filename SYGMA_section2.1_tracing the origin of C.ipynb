This is the second teaching of SYGMA, in this section, they try to find out the origin of C

```python
import matplotlib.pyplot as plt
from NuPyCEE import sygma as s
from NuPyCEE import omega as o
from NuPyCEE import stellab as e
```

```python
stellab=e.stellab()
```

The parameters are for Milky Way

```python
o_mw=o.omega(iniZ=0.0001, galaxy='milky_way', Z_trans=-1, 
             table='yield_tables/agb_and_massive_stars_nugrid_MESAonly_fryer12delay.txt',
             sfe=0.005, DM_evolution=True, mass_loading=1, nb_1a_per_m=3e-3, special_timesteps=60)
```

New parameters: galaxy, Z_trans, sfe, DM_evolution, mass_loading, nb_1a_per_m

In the model stellab, the data is from the observation. We compare the reason of OMEGA with the observations.

```python
# set the xaxis and yaxis
x='[Fe/H]'
y='[C/Fe]'

# First, plot the observational data by the STELLAB
stellab.plot_spectro(fig=1, xaxis=x, yaxis=y, norm='Grevesse_noels_1993',galaxy='milky_way',show_err=False)

# Then plot the numerical predictions of OMEGA
xy_f=o_mw.plot_spectro(fig=1, xaxis=x, yaxis=y, return_x_y=True)
plt.plot(xy_f[0],xy_f[1],linewidth=4,color='w')
plt.plot(xy_f[0],xy_f[1],linewidth=2,color='k',label='OMEGA')

# Set the parameters of plot
plt.legend(loc='center left', bbox_to_anchor=(1.01,0.5),markerscale=0.8,fontsize=13)
plt.xlim(-4,0.5)
plt.ylim(-1.4,1.6)
```

### For the SSP

Try to set the first stellar popualtion in the galaxy, they formed from Z=0.0001

```python
sp0001=s.sygma(iniZ=0.0001)
sp001=s.sygma(iniZ=0.001)
```

The C abundance of these two populations

```python
element='[C/Fe]'
sp0001.plot_spectro(fig=2,yaxis=element,marker='D',color='b',label='Z=0.0001')
sp001.plot_spectro(fig=2,yaxis=element,label='0.001')
```

```python
# The ejected mass of a certain element
sp0001.plot_mass(fig=3,specie='C',marker='D',color='b',label='Z=0.0001')
sp001.plot_mass(fig=3,specie='C',label='Z=0.001')
```

The different contribution of different stars.

```python
sp0001.plot_mass_range_contributions(specie='C',marker='D',color='b',label='Z=0.0001')
sp001.plot_mass_range_contributions(specie='C',label='Z=0.001')
```
