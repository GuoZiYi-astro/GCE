```python
# python packages
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from NuPyCEE import omega as o
```

## Galactic inflows

### Galactic inflows and appearance of SNe Ia

```python
# simulate the effect of the SNe Ia
# set the constant SFR
# when the mgal=1e9 the gas is not enough to sustain the SFH

o_11_close=o.omega(cte_sfr=1,mgal=1e11)
o_10_close=o.omega(cte_sfr=1,mgal=1e10)
o_9_close=o.omega(cte_sfr=1,mgal=1e9)

o_11_close_noIa=o.omega(cte_sfr=1,mgal=1e11,sn1a_on=False)
o_10_close_noIa=o.omega(cte_sfr=1,mgal=1e10,sn1a_on=False)
o_9_close_noIa=o.omega(cte_sfr=1,mgal=1e9,sn1a_on=False)
```

```python
# compare the result. 
y='[C/Fe]'
x='[Fe/H]'

o_11_close.plot_spectro(fig=1,xaxis=x,yaxis=y,color='r',label='mgal=1e11')
o_10_close.plot_spectro(fig=1,xaxis=x,yaxis=y,color='g',label='mgal=1e10')
o_9_close.plot_spectro(fig=1,xaxis=x,yaxis=y,color='b',label='mgal=1e9')

o_11_close_noIa.plot_spectro(fig=2,xaxis=x,yaxis=y,color='r',label='mgal=1e11',shape='--')
o_10_close_noIa.plot_spectro(fig=2,xaxis=x,yaxis=y,color='g',label='mgal=1e10',shape='--')
o_9_close_noIa.plot_spectro(fig=2,xaxis=x,yaxis=y,color='b',label='mgal=1e9',shape='--')

o_11_close.plot_spectro(fig=3,xaxis=x,yaxis=y,color='r',label='mgal=1e11')
o_11_close_noIa.plot_spectro(fig=3,xaxis=x,yaxis=y,color='r',label='mgal=1e11,no SNe Ia',shape='--')
```
