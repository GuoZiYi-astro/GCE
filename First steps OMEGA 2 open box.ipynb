```python
# python packages
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from NuPyCEE import omega as o
```

## Galactic inflows

### Galactic inflows and appearance of SNe Ia

```python
# simulate the effect of the SNe Ia
# set the constant SFR
# when the mgal=1e9 the gas is not enough to sustain the SFH

o_11_close=o.omega(cte_sfr=1,mgal=1e11)
o_10_close=o.omega(cte_sfr=1,mgal=1e10)
o_9_close=o.omega(cte_sfr=1,mgal=1e9)

o_11_close_noIa=o.omega(cte_sfr=1,mgal=1e11,sn1a_on=False)
o_10_close_noIa=o.omega(cte_sfr=1,mgal=1e10,sn1a_on=False)
o_9_close_noIa=o.omega(cte_sfr=1,mgal=1e9,sn1a_on=False)
```

```python
# compare the result. 
y='[C/Fe]'
x='[Fe/H]'

o_11_close.plot_spectro(fig=1,xaxis=x,yaxis=y,color='r',label='mgal=1e11')
o_10_close.plot_spectro(fig=1,xaxis=x,yaxis=y,color='g',label='mgal=1e10')
o_9_close.plot_spectro(fig=1,xaxis=x,yaxis=y,color='b',label='mgal=1e9')

o_11_close_noIa.plot_spectro(fig=2,xaxis=x,yaxis=y,color='r',
                             label='mgal=1e11',shape='--')
o_10_close_noIa.plot_spectro(fig=2,xaxis=x,yaxis=y,color='g',
                             label='mgal=1e10',shape='--')
o_9_close_noIa.plot_spectro(fig=2,xaxis=x,yaxis=y,color='b',
                            label='mgal=1e9',shape='--')

o_11_close.plot_spectro(fig=3,xaxis=x,yaxis=y,color='r',label='mgal=1e11')
o_11_close_noIa.plot_spectro(fig=3,xaxis=x,yaxis=y,color='r',
                             label='mgal=1e11,no SNe Ia',shape='--')

o_10_close.plot_spectro(fig=4,xaxis=x,yaxis=y,color='g',label='mgal=1e10')
o_10_close_noIa.plot_spectro(fig=4,xaxis=x,yaxis=y,color='g',
                             label='mgal=1e10,no SNe Ia',shape='--')

o_9_close.plot_spectro(fig=5,xaxis=x,yaxis=y,color='b',label='mgal=1e9')
o_9_close_noIa.plot_spectro(fig=5,xaxis=x,yaxis=y,color='b',
                            label='mgal=1e9,no SNe Ia',shape='--')
```

In Milky Way, the SN Ia appear around [Fe/H]=-1, so in the closed box, the Milky Way seems like have a low intial mass of gas.

```python
# set the inflow and outflow with or without SNe Ia

o_9_open=o.omega(mgal=1e9,cte_sfr=1,in_out_control=True,outflow_rate=0,inflow_rate=1.0)

o_9_open_no1a=o.omega(mgal=1e9,cte_sfr=1,in_out_control=True,outflow_rate=0,
                      inflow_rate=1,sn1a_on=False)
```

```python
# still plot the same figure
x='[Fe/H]'
y='[C/Fe]'

o_9_open.plot_spectro(fig=6,xaxis=x,yaxis=y,label='mgal=1e9',color='b')
o_9_open_no1a.plot_spectro(fig=6,xaxis=x,yaxis=y,label='mgal=1e9, no SNe Ia')
```

### Dilution process

```python
# simulate constant SFR and different inflow rate.

o_in_0=o.omega(mgal=2e10,cte_sfr=1.0,in_out_control=True,outflow_rate=0,inflow_rate=0)
o_in_1=o.omega(mgal=2e10,cte_sfr=1.0,in_out_control=True,outflow_rate=0,inflow_rate=1)
o_in_3=o.omega(mgal=2e10,cte_sfr=1.0,in_out_control=True,outflow_rate=0,inflow_rate=3)
o_in_9=o.omega(mgal=2e10,cte_sfr=1.0,in_out_control=True,outflow_rate=0,inflow_rate=9)
```

```python
a='H'
b='Ca'
c='Fe'

o_in_0.plot_mass(fig=7,specie=a,color='k',label='inflow_rate=0.0 Msun/yr')
o_in_1.plot_mass(fig=7,specie=a,color='r',label='inflow_rate=1.0 Msun/yr')
o_in_3.plot_mass(fig=7,specie=a,color='b',label='inflow_rate=3.0 Msun/yr')
o_in_9.plot_mass(fig=7,specie=a,color='g',label='inflow_rate=9.0 Msun/yr')

o_in_0.plot_mass(fig=7,specie=b,color='k',show_legend=False)
o_in_1.plot_mass(fig=7,specie=b,color='r',show_legend=False)
o_in_3.plot_mass(fig=7,specie=b,color='b',show_legend=False)
o_in_9.plot_mass(fig=7,specie=b,color='g',show_legend=False)

o_in_0.plot_mass(fig=7,specie=c,color='k',show_legend=False)
o_in_1.plot_mass(fig=7,specie=c,color='r',show_legend=False)
o_in_3.plot_mass(fig=7,specie=c,color='b',show_legend=False)
o_in_9.plot_mass(fig=7,specie=c,color='g',show_legend=False)

plt.xscale('linear')
plt.ylim(1e3,3e11)

plt.annotate('H',  xy=(0.2e10, 3.0e9))
plt.annotate('Ca', xy=(1.0e10, 1.5e5))
plt.annotate('Fe', xy=(1.0e10, 3.0e7))
```

```python
# The evolution of [Fe/H]
y='[Fe/H]'
o_in_0.plot_spectro(fig=8,yaxis=y,color='r',label='inflow_rate=0.0 Msun/yr')
o_in_1.plot_spectro(fig=8,yaxis=y,color='b',label='inflow_rate=0.0 Msun/yr')
o_in_3.plot_spectro(fig=8,yaxis=y,color='g',label='inflow_rate=0.0 Msun/yr')
o_in_9.plot_spectro(fig=8,yaxis=y,color='m',label='inflow_rate=0.0 Msun/yr')
plt.ylim(-3,0.5)
```

```python
# The effect of inflow for different stars
a='Fe'

o_in_0.plot_mass(fig=9,specie=a,color='k',label='inflow_rate=0.0 Msun/yr',source='massive')
o_in_1.plot_mass(fig=9,specie=a,color='r',label='inflow_rate=1.0 Msun/yr',source='massive')
o_in_3.plot_mass(fig=9,specie=a,color='b',label='inflow_rate=3.0 Msun/yr',source='massive')
o_in_9.plot_mass(fig=9,specie=a,color='g',label='inflow_rate=9.0 Msun/yr',source='massive')
plt.annotate('massive stars',xy=(5e8,6))

o_in_0.plot_mass(fig=10,specie=a,color='k',label='inflow_rate=0.0 Msun/yr',source='agb')
o_in_1.plot_mass(fig=10,specie=a,color='r',label='inflow_rate=1.0 Msun/yr',source='agb')
o_in_3.plot_mass(fig=10,specie=a,color='b',label='inflow_rate=3.0 Msun/yr',source='agb')
o_in_9.plot_mass(fig=10,specie=a,color='g',label='inflow_rate=9.0 Msun/yr',source='agb')
plt.annotate('agb stars',xy=(5e6,1e3))

o_in_0.plot_mass(fig=11,specie=a,color='k',label='inflow_rate=0.0 Msun/yr',source='sn1a')
o_in_1.plot_mass(fig=11,specie=a,color='r',label='inflow_rate=1.0 Msun/yr',source='sn1a')
o_in_3.plot_mass(fig=11,specie=a,color='b',label='inflow_rate=3.0 Msun/yr',source='sn1a')
o_in_9.plot_mass(fig=11,specie=a,color='g',label='inflow_rate=9.0 Msun/yr',source='sn1a')
plt.annotate('sn1a',xy=(5e6,1e3))
```

From these figures, we can think that the inflow rate will change the agb stars to influence the yields.

## Star formation with inflows and outflows

### Mass of gas with the SF_law parameter

```python
# set the SF_law with constant SFR
o_nosf=o.omega(cte_sfr=1,mgal=1e10)
o_sf=o.omega(cte_sfr=1,SF_law=True)
```

```python
o_nosf.plot_totmasses(fig=17,color='r',label='Closed box')
o_sf.plot_totmasses(fig=17,color='b',label='SF_law=True')
plt.ylim(8e8,2e10)
```

$$M_{gas}(t)=\frac{\tau}{sfe}SFH(t)$$

```python
o_sf.plot_inflow_rate(fig=18,color='g',shape='--')
```

```python
# set different SFE
o_consf_sfe01=o.omega(cte_sfr=1,SF_law=True,sfe=0.01)
o_consf_sfe04=o.omega(cte_sfr=1,SF_law=True,sfe=0.04)
o_consf_sfe16=o.omega(cte_sfr=1,SF_law=True,sfe=0.16)
```

```python
#Plot the mass of gas reservoir
o_consf_sfe01.plot_totmasses(fig=19,color='r',label='sfe=0.01')
o_consf_sfe04.plot_totmasses(fig=19,color='g',label='sfe=0.04')
o_consf_sfe16.plot_totmasses(fig=19,color='b',label='sfe=0.16')
```

```python
o_consf_sfe01.plot_inflow_rate(fig=20,color='r',label='sfe=0.01')
o_consf_sfe04.plot_inflow_rate(fig=20,color='g',label='sfe=0.04')
o_consf_sfe16.plot_inflow_rate(fig=20,color='b',label='sfe=0.16')
```

```python
o_consf_sfe01.plot_outflow_rate(fig=21,color='r',label='sfe=0.01')
o_consf_sfe04.plot_outflow_rate(fig=21,color='g',label='sfe=0.04',shape='--')
o_consf_sfe16.plot_outflow_rate(fig=21,color='b',label='sfe=0.16',shape='-.')
```

For the SF_law=True mode, the outflow_rate is constant after the star start to formation, while the inflow_rate change with time.

### Galactic inflows and outflows

$$M_{out}(t)=\eta\, SFH(t).$$

```python
# set different mass_loading
o_sf_ml_0=o.omega(cte_sfr=1,SF_law=True,sfe=0.1,mass_loading=0)
o_sf_ml_1=o.omega(cte_sfr=1,SF_law=True,sfe=0.1,mass_loading=1)
o_sf_ml_3=o.omega(cte_sfr=1,SF_law=True,sfe=0.1,mass_loading=3)
o_sf_ml_9=o.omega(cte_sfr=1,SF_law=True,sfe=0.1,mass_loading=9)
```
